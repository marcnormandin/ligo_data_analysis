```python
# Plot the PTAPPSO results
import numpy as np
import math
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
ra, dec, snr = np.genfromtxt('snr_sky_map.dat').transpose()
```

```python
# see: http://stackoverflow.com/questions/27004422/contour-imshow-plot-for-irregular-x-y-z-data
plt.figure(figsize=(9,5))
plt.tricontourf(ra, dec, snr, 40)

# add the true source location (we are using simulated data)
source_ra = -2.14
source_dec = 0.72
plt.plot(source_ra, source_dec, 'ko', markersize=10, markerfacecolor="None")

plt.xlabel('Right-ascension')
plt.ylabel('Declination')
plt.title('Network SNR Sky Map')

plt.colorbar()

plt.tight_layout()

plt.savefig('snr_sky_map.png')
```
